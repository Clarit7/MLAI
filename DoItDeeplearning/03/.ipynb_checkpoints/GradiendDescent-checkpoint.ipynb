# 경사하강법
## 예측값으로 올바른 모델 찾기
### w와 b 초기화하기

```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

x = diabetes.data[:, 2]
y = diabetes.target

w = 1.0
b = 1.0
```

### 훈련 데이터의 첫 번째 샘플 데이터로 y_hat 얻기

```python
y_hat = x[0] * w + b
print(y_hat)
```

### w 값 조절해 예측값 바꾸기

```python
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
print(y_hat_inc)
```

### w 값 조정한 후 예측값 증가 정도 확인하기

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```

## 변화율로 가중치 업데이트 하기

```python
w_new = w + w_rate
print(w_new)
```

## 변화율로 절편 업데이트 하기

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)

b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)

b_new = b + b_rate
print(b_new)
```

## 오차 역전파로 가중치와 절편 업데이트
### 오차와 변화율을 곱하여 가중치 업데이트

```python
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

### 두 번째 샘플 x[1]을 사용해 오차를 구하고 새로운 w와 b 구하기

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

### 전체 샘플을 반복하기

```python
for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err
print(w, b)
```

### 그래프로 모델 확인

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 여러 에포크를 반복하기

```python
for i in range(1, 100):
    for x_i, y_i in zip(x, y):
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate * err
        b = b + 1 * err
print(w, b)

plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 모델로 예측하기

```python
x_new = 0.18
y_pred = x_new * w + b
print(y_pred)

plt.scatter(x, y)
plt.scatter(x_new, y_pred)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
