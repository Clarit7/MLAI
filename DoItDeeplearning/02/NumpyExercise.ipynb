# 파이썬 기초 익히기

## 1차원 배열 생성

```python
my_list = [10, 'hello list', 20]
print(my_list[1])
```

## 2차원 배열 생성

```python
my_list_2 = [[10, 20, 30], [40, 50, 60]]
print(my_list_2[1][1])
```

## NumPy
### NumPy import

```python
import numpy as np
print(np.__version__)
```

### array() 함수로 2차원 배열 만들기

```python
my_arr = np.array([[10, 20, 30], [40, 50, 60]])
print(my_arr)
```

### type() 함수로 넘파이 배열인지 확인하기

```python
type(my_arr)
```

### 넘파이 배열에서 요소 선택하기

```python
my_arr[0][2]
```

### 넘파이 내장 함수 사용하기

```python
np.sum(my_arr)
```

## Matplotlib
### Matplotlib import

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 선 그래프 그리기

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

### 산점도 그리기

```python
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

### 넘파이 배열로 산점도 그리기

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```
