# Anomaly Detection

## Loading the Data

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from tqdm.notebook import tqdm
```

```python
mat = loadmat("ex8data1.mat")
X = mat["X"]
Xval = mat["Xval"]
yval = mat["yval"]
```

## Plotting the data

```python
plt.scatter(X[:,0],X[:,1],marker="x")
plt.xlim(0,30)
plt.ylim(0,30)
plt.xlabel("Latency (ms)")
plt.ylabel("Throughput (mb/s)")
plt.show()
```

## Gaussian Distribution

$p(x;\mu,\sigma^2)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{\frac{(x-\mu)^2}{2\sigma^2}}$

$\mu_i=\frac{1}{m}\sum^{m}_{j=1}x^{(j)}$

$\sigma_i^2=\frac{1}{m}\sum^{m}_{j=1}(x^{(j)}-\mu_j)^2$

```python
def estimateGaussian(X):
    m = X.shape[0]
    
    # mean
    sum_ = np.sum(X,axis=0)
    mu = 1/m *sum_
    
    # variance
    var = 1/m * np.sum((X - mu)**2,axis=0)
    
    return mu,var
```

```python
mu, sigma2 = estimateGaussian(X)
```

## Multivariate Gaussian Distribution

$p(x;\mu,\Sigma)=\frac{1}{(2\pi)^{n/2}|\Sigma|^{1/2}}exp(-\frac{1}{2}(x-\mu)^T\Sigma^{-1}(x-\mu))$

numpy.linalg.det() : 행렬식(determinant) 계산  
np.linalg.pinv() : SVD를 이용한 pseudo-inverse matrix 계산

```python
def multivariateGaussian(X, mu, sigma2):
    k = len(mu)
    
    sigma2=np.diag(sigma2)
    X = X - mu
    p = 1/((2*np.pi)**(k/2)*(np.linalg.det(sigma2)**0.5))* np.exp(-0.5* np.sum(X @ np.linalg.pinv(sigma2) * X,axis=1))
    return p
```

```python
p = multivariateGaussian(X, mu, sigma2)
```

## Visualize the fit

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],marker="x")
X1,X2 = np.meshgrid(np.linspace(0,35,num=70),np.linspace(0,35,num=70))
p2 = multivariateGaussian(np.hstack((X1.flatten()[:,np.newaxis],X2.flatten()[:,np.newaxis])), mu, sigma2)
contour_level = 10**np.array([np.arange(-20,0,3,dtype=np.float)]).T

"""""
원본 소스코드엔 파라미터 levels에 contour_level이 입력돼있음
flatten()함수를 통해 1차원 array로 바꿔줘야 정상적으로 contour 함수가 작동
"""""
plt.contour(X1,X2,p2[:,np.newaxis].reshape(X1.shape),contour_level.flatten())

plt.xlim(0,35)
plt.ylim(0,35)
plt.xlabel("Latency (ms)")
plt.ylabel("Throughput (mb/s)")
plt.show()
```

## Selecting the threshold $\epsilon$

```python
def selectThreshold(yval, pval):
    best_epi = 0
    best_F1 = 0
    
    """
    pval의 최소값과 최대값을 1000개의 구간으로 나누고
    가장 높은 F1score를 보이는 적절한 epsilon을 고름
    """
    stepsize = (max(pval) -min(pval))/1000
    epi_range = np.arange(pval.min(),pval.max(),stepsize)
    
    for epi in tqdm(epi_range):
        predictions = (pval<epi)[:,np.newaxis]
        tp = np.sum(predictions[yval==1]==1)
        fp = np.sum(predictions[yval==0]==1)
        fn = np.sum(predictions[yval==1]==0)
        
        # 0으로 나누는 오류 방지
        if tp+fp != 0:
            prec = tp/(tp+fp)
            rec = tp/(tp+fn)

            F1 = (2*prec*rec)/(prec+rec)

            if F1 > best_F1:
                best_F1 =F1
                best_epi = epi
        
    return best_epi, best_F1
```

```python
pval = multivariateGaussian(Xval, mu, sigma2)
epsilon, F1 = selectThreshold(yval, pval)
print("Best epsilon found using cross-validation:",epsilon)
print("Best F1 on Cross Validation Set:",F1)
```

## Visualize the anomalies

numpy.nonzero : array에서 0이 아닌 값들의 index 반환

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],marker="x")
X1,X2 = np.meshgrid(np.linspace(0,35,num=70),np.linspace(0,35,num=70))
p2 = multivariateGaussian(np.hstack((X1.flatten()[:,np.newaxis],X2.flatten()[:,np.newaxis])), mu, sigma2)
contour_level = 10**np.array([np.arange(-20,0,3,dtype=np.float)]).T
plt.contour(X1,X2,p2[:,np.newaxis].reshape(X1.shape),contour_level.flatten())

outliers = np.nonzero(p<epsilon)[0]
plt.scatter(X[outliers,0],X[outliers,1],marker ="o",facecolor="none",edgecolor="r",s=70)

plt.xlim(0,35)
plt.ylim(0,35)
plt.xlabel("Latency (ms)")
plt.ylabel("Throughput (mb/s)")
plt.show()
```
